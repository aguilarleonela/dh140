```python
def logli(word):
    a = romeoandjulietfreq[word]
    b = completefreq[word]
    c = len(romeoandjulietstemmed)
    d = len(completestemmed)
    E1 = c*(a+b)/(c+d)
    E2 = d*(a+b)/(c+d)
    if a != 0 and b != 0:
        G2 = 2*((a*np.log(a/E1)) + (b*np.log(b/E2)))
    else:
        G2 = 0
    return G2
```

```python
completefreq['love']/len(completestemmed)
```

```python
wordlist = []
wordfreq = []
loglike = []
for i in sorted(romeoandjulietfreq, key=romeoandjulietfreq.get, reverse=True):
    specialword = logli(i)
    if specialword > -1:
        wordlist.append(i)
        wordfreq.append(romeoandjulietfreq[i])
        loglike.append(specialword)
df = pd.DataFrame({'wordlist':wordlist,'wordfreq':wordfreq,'loglike':loglike})
```
