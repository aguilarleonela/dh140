#### These are the characters from the play 

```python
characters = [
    'Romeo',
    'Juliet',
    'Nurse',
    'Tybalt',
    'Lady Capulet',
    'Friar Lawrence',
    'Mercutio',
    'Capulet',
    'Montague',
    'Paris',
    'Benvolio',
    'Prince Escalus',
    'Friar John',
    'Balthasar',
    'Sampson',
    'Gregory',
    'Abram',
    'Apothecary',
    'Peter',
    'Rosaline',
    'Chorus',
   
]
```

```python
len(acts)
```

```python
charnum = {}

connections = {}
for i in range(len(characters)-1):
    for j in range(i+1,len(characters)):
        connections[(characters[i],characters[j])] = 0

for k in characters:
    charnum[k] = 0

for i in acts:
    for j in i.split('scene')[0:]:
        scenechars = []
        for k in characters:
            if j.find(k) != -1:
                scenechars.append(k)
                charnum[k] += 1
        for a in range(len(scenechars)-1):
            for b in range(a+1,len(scenechars)):
                connections[(scenechars[a],scenechars[b])] += 1
```

#### The degree centrality measures the 'importance' of node by assigning a score based on the number of bridges held by each node.

```python
nx.degree_centrality(G)
```

```python
dc = nx.degree_centrality(G)
for i in sorted(dc, key=dc.get, reverse=True):
    print('{:15s}: {:.3f}'.format(i.title(), dc[i]))
```

```python
from networkx.algorithms.community.centrality import girvan_newman
```

```python
communities = girvan_newman(G)
```

```python
node_groups = []
for com in next(communities):
    node_groups.append(list(com))
 
print(node_groups)
 
color_map = []
for node in G:
    if node in node_groups[0]:
        color_map.append('blue')
    else:
        color_map.append('green')
nx.draw(G, node_color=color_map, with_labels=True)
plt.show()
```

#### The information in this graph displays the network of the play among the characters. Each node has a bridge which distinguishes their relationships amongst each other. For example, Romeo has the most bridges and edges among the characters. Evidently, the edges of Romeo showcase  metrics of connectedness and centralization. We can also observe the the degree of nearness (direct or indirect) between the node and the rest of the nodes. Benvolio for example, is a node that is distant from the rest with the least amount of edges. The density of the network according to the code is .92. The most important nodes are the ones with the highest closeness centrality and degree centrality. These factors help identify significant actors in a social network. Degree centrality is a node with a great number of links and connections in a network. Both Romeo and Balthasar have strong networks in the graph. Between centrality also plays a role in the importance of nodes. Nodes with high betweeness centrality help information flow, especially in cohesive networks like this one.

```python
from networkx.algorithms.community import greedy_modularity_communities
```

```python
greedy_modularity_communities(G)
```

```python
communities = greedy_modularity_communities(G)
# Create empty dictionary
modularity_class = {}
#Loop through each community in the network
for community_number, community in enumerate(communities):
    #For each member of the community, add their community number
    for name in community:
        modularity_class[name] = community_number
```

```python
modularity_class
```

```python
communities = greedy_modularity_communities(G)

node_groups = []
for com in communities:
    node_groups.append(list(com))
 
print(node_groups)
 
color_map = []
for node in G:
    if node in node_groups[0]:
        color_map.append('blue')
    elif node in node_groups[1]:
        color_map.append('yellow')
    else:
        color_map.append('green')
nx.draw(G, node_color=color_map, with_labels=True)
plt.show()
```
